```python
'''
rename the Unnamed column to DepMap_ID as the column contains the DepMap ID's of the cell lines 
These DepMap_ID's help in merging the ccle data with the sample_info data
'''

ccle.rename(columns={'Unnamed: 0':'DepMap_ID'}, inplace=True)
```
